```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import itertools

# List of file paths
file_paths = ['TITAN.CSV', 'RELIANCE.CSV', 'ASIANPAINT.CSV', 'HEROMOTOCO.CSV', 'L&T.csv', 'TCS.CSV']

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each file and append its DataFrame to the list
for file_path in file_paths:
    df = pd.read_csv(file_path)
    # Assuming 'Date' is a column in your DataFrame
    df['Date'] = pd.to_datetime(df['Date'])
    dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
data = pd.concat(dfs, ignore_index=True)

# Create a color map for each stock
colors = iter(plt.cm.rainbow(np.linspace(0, 1, len(file_paths))))

# Create a list to store colors for individual line charts
line_chart_colors = []

plt.figure(figsize=(28, 14))

for file_path in file_paths:
    df = pd.read_csv(file_path)
    color = next(colors)
    line_chart_colors.append(color)
    plt.plot(df['Date'], df['Close'], label=file_path, color=color)

plt.title(f'Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()

# Plot individual line charts with the same color as the combined chart
num_rows = 3
num_cols = 2
num_charts = min(len(file_paths), num_rows * num_cols)

for i in range(num_charts):
    plt.figure(figsize=(20, 15))
    for j in range(num_rows):
        for k in range(num_cols):
            chart_index = i * num_rows * num_cols + j * num_cols + k
            if chart_index < len(file_paths):
                file_path = file_paths[chart_index]
                df = pd.read_csv(file_path)
                color = line_chart_colors[chart_index]
                plt.subplot(num_rows, num_cols, j * num_cols + k + 1)
                plt.plot(df['Date'], df['Close'], label=file_path, color=color)
                plt.title(f'Closing Prices Over Time - {file_path}')
                plt.xlabel('Date')
                plt.ylabel('Closing Price')
                plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap

# List of file paths
file_paths = ['TITAN.CSV', 'RELIANCE.CSV', 'ASIANPAINT.CSV', 'HEROMOTOCO.CSV', 'L&T.csv', 'TCS.CSV']
# Initialize an empty list to store DataFrames
dfs = []

# Define colors for each stock
stock_colors = {
    'TITAN.CSV': '#86C8BC',
    'RELIANCE.CSV': '#C07F00',
    'ASIANPAINT.CSV': '#E36414',
    'HEROMOTOCO.CSV': 'red',
    'L&T.csv': '#11235A',
    'TCS.CSV': '#DA0C81'
}

for file_path in file_paths:
    try:
        # Read each CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Convert 'Date' column to datetime
        df['Date'] = pd.to_datetime(df['Date'])
        
        # Calculate the percentage change for the closing price
        df['Percentage Change'] = ((df['Close'] - df['Close'].iloc[0]) / df['Close'].iloc[0]) * 100
        
        # Extract the year from the 'Date' column
        df['Year'] = df['Date'].dt.year
        
        # Add a new column with the file name
        df['File'] = file_path
        
        # Append the DataFrame to the list
        dfs.append(df)
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
data = pd.concat(dfs, ignore_index=True)

# Create a 3x3 grid of subplots
fig, axes = plt.subplots(12, 3, figsize=(20, 40), sharex=True, sharey=True)

# Flatten the 3x3 array of subplots
axes = axes.flatten()

# Loop through each stock and create a heatmap for each year
for i, (file_path, subset) in enumerate(data.groupby(['File', 'Year'])):
    file, year = file_path
    subset_pivot = subset.pivot(index='Date', columns='File', values='Percentage Change')
    
    # Get the color for the current stock
    color = stock_colors[file]
    
    # Create a custom colormap for the stock
    cmap = ListedColormap(sns.color_palette([color]))

    # Create a heatmap with custom colors
    sns.heatmap(subset_pivot.resample('Y').last(), ax=axes[i], cmap=cmap,   annot=True, fmt=".2f")

    axes[i].set_title(f"{file.split('.')[0]} - {year}")  # Set subplot title


# Adjust layout
plt.tight_layout()
plt.show()
```
